Extract Function

```python
def extract_csv(path):
    import csv
    with open(path,'r') as out_file:
       return list(csv.DictReader(out_file))
 
```

Parse Function

```python
def parse_rows(rows):
    parsed = []

    for row in rows:
        parsed.append({
            "product_id":int(row['product_id']),
            "price": float(row['price']),
            "product_name": row['product_name'].strip().title(),
            "category": row['category'].lower()
            })
    
    return parsed
```

Validate function

```python
def validate_rows_(rows):
    valid = []

    for row in rows:

        if row['price'] <= 0:
            continue
        if not row['product_name']:
            continue
        
        valid.append(row)
    return valid
```

Transform Function

```python
def transform(rows):
    transform_data = []

    for row in rows:
        r2 = row.copy() #avoid mutation
        r2['price_with_tax'] = r2['price']*1.2
    
        
        transform_data.append(row)
    
    return transform_data
```

Load Function

```python
def load_csv(path,rows):

    with open(path,'w',newline='') as out_file:

        writer = csv.DictWriter(out_file, fieldnames=rows[0].keys())
        writer.writeheader()
        writer.writerows(rows)
```

Run Pipeline Function

```python
def run_pipeline(input_path, output_path):

    raw = extract_csv(input_path)
    parsed = parse_rows(raw)
    validated = validate_rows_(parsed)
    transformed=transform(validated)
    load_csv(output_path,transformed)

    return transformed
```

```python
import csv
from datetime import datetime, date, timedelta
import copy


data = [
    {'sale_id': 1, 'amount_usd': 100, 'rates_type': 'standard'},
    {'sale_id': 2, 'amount_usd': 200, 'rates_type': 'standard'},
    {'sale_id': 3, 'amount_usd': 300, 'rates_type': 'premium'},
    {'sale_id': 4, 'amount_usd': 150, 'rates_type': 'premium'},
    {'sale_id': 5, 'amount_usd': -50, 'rates_type': 'standard'}
]
with open('sales.csv','w',newline = '') as out_file:

    writer = csv.DictWriter(out_file, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)


def extract_csv(path):
    with open(path,'r') as out_file:
        return list(csv.DictReader(out_file))

def parse_rows(rows):
    parsed = []
    for row in rows:
        parsed.append({
        'sale_id': int(row['sale_id']),
        'amount_usd': float(row['amount_usd']),
        'rates_type': row['rates_type']
        })
    return parsed

def validate_rows(rows):
    valid= []

    for row in rows:
        if row['amount_usd'] < 0:
            continue
        
        valid.append(row)
    return valid

def transform_rows(rows):
    transform = []

    new_rows = copy.deepcopy(rows)
    for r2 in new_rows:
        
        if r2['rates_type'] == 'standard':
        #    calculated_value = r2['amount_usd'] *1.1
        #    calculated_value = round(calculated_value,1)
        #    r2['amount_converted'] = calculated_value
            r2['amount_converted'] = round(r2['amount_usd'] * 1.1, 1)   
        elif r2['rates_type']== 'premium':
             r2['amount_converted'] = round(r2['amount_usd'] * 1.2, 1)
        
        transform.append(r2)
    
    return transform

def load_csv(path, rows):
    with open(path,'w',newline='') as out_file:

        writer = csv.DictWriter(out_file, fieldnames=rows[0].keys())
        writer.writeheader()
        writer.writerows(rows)


def run_pipeline(input_path, output_path):

    raw = extract_csv(input_path)
    parsed = parse_rows(raw)
    validated = validate_rows(parsed)
    transformed=transform_rows(validated)
    load_csv(output_path,transformed)

    return transformed


```

```python
def process_row(row):
    row["id"] = int(row["id"])
    if row["amount"] == "" or row["amount"] is None or float(row["amount"]) < 0:
        return None
    row["amount"] = float(row["amount"])
    row["amount_tax"] = round(row["amount"] * 1.2, 2)
    return row
```

**Rewriting the above cleanly applying these prinicples KISS, SRP**


```python
def parse_rows(rows):
    parsed = []

    for row in rows:
        r_copy = row.copy()
        r_copy["id"] = int(r_copy["id"])
        r_copy["amount"] = float(r_copy["amount"])
        parsed.append(r_copy)
    
    return parsed
```

```python
def validate_rows(rows):
    valid  = []

    for row in rows:
         if row["amount"] >= 0:
              valid.append(row)
    return valid
 
```

```python
def transform_rows(rows):
    output = []

    for row in rows:
        r_copy = row.copy()
        r_copy["amount_tax"] = round(r_copy["amount"] * 1.2, 2)

        output.append(r_copy)
    return output
```
